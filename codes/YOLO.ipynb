```python
def find_latest_model(folder):
    best_model = None
    latest_time = 0
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file == "best.pt":
                full_path = os.path.join(root, file)
                mod_time = os.path.getmtime(full_path)
                if mod_time > latest_time:
                    latest_time = mod_time
                    best_model = full_path
    return best_model

chunks_folder = r"C:/Users/Begad/Documents/training/final project/chunk"
model_path = find_latest_model(chunks_folder)
print("Using model:", model_path)
```
